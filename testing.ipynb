```python
def filter_signal(cell_number, tracked_masks, images, T):
    intensities = get_cell_intensities(cell_number, tracked_masks, images, T, normalize=False)
    intensities[0] = np.median(intensities[0:1])
    for i in range(1,len(intensities)):
        intensities[i] = np.median(intensities[i-1:i+1])
    intensities = intensities/np.max(intensities)
    mean = np.median(intensities)
    for i in range(len(intensities)):
        if intensities[i]<mean:
            intensities[i] = 0
    return intensities
```

```python
def filter_signal2(cell_number, tracked_masks, images, T, n=None, hpf=0.0005, lpf=0.017, cutoff_peakheight=None):
    intensities = get_cell_intensities(cell_number, tracked_masks, images, T, normalize=False)
    if n:
        for i in range(n):
            intensities[i] = np.mean(intensities[0:i+n])
        for i in range(4,len(intensities)):
            intensities[i] = np.mean(intensities[i-n:i+n])
    if hpf:
        hpf_cutoff_freq = hpf
        freqs = np.fft.fftfreq(len(intensities), T)
        filter_mask = np.abs(freqs) > hpf_cutoff_freq
        intensities_fft = np.fft.fft(intensities)
        filtered_signal = np.real(np.fft.ifft(intensities_fft*filter_mask))
        intensities = filtered_signal
    if lpf:
        lpf_cutoff_freq = lpf
        freqs = np.fft.fftfreq(len(intensities), T)
        filter_mask = np.abs(freqs) < lpf_cutoff_freq
        intensities_fft = np.fft.fft(intensities)
        filtered_signal = np.real(np.fft.ifft(intensities_fft*filter_mask))
        intensities = filtered_signal
    intensities = intensities/np.max(intensities)
    if cutoff_peakheight!=None:
        for i in range(len(intensities)):
            if intensities[i]<cutoff_peakheight:
                intensities[i] = 0
    return intensities
```

```python
def correlation(intensities1, intensities2, max_diff):
    def CC(f1, f2):
        return np.sum(f1*f2)
    
    def CCN(f1, f2):
        return CC(f1,f2)/np.sqrt(CC(f1,f1)*CC(f2,f2))
    
    correlation_list = []
    for dt in range(max_diff,0,-1):
        correlation_list.append(CCN(intensities1[:-dt],intensities2[dt:]))
    correlation_list.append(CCN(intensities1,intensities2))
    for dt in range(1, max_diff+1):
        correlation_list.append(CCN(intensities1[dt:],intensities2[:-dt]))

    return correlation_list
```

```python
def correlation(intensities1, intensities2, max_diff):
    def CC(f1, f2):
        return np.sum(f1*f2)
    
    def CCN(f1, f2):
        return CC(f1,f2)/np.sqrt(CC(f1,f1)*CC(f2,f2))
    
    correlation_list = []

    correlation_list.append(CCN(intensities1,intensities2))
    for dt in range(1, max_diff):
        correlation_list.append(CCN(intensities2[dt:],intensities1[:-dt]))
    correlation_list.reverse()
    for dt in range(1, max_diff):
        correlation_list.append(CCN(intensities1[dt:],intensities2[:-dt]))

    return correlation_list
```
